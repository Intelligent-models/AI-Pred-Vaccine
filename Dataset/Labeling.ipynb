```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv(r'preprocessed.csv')
data
```

```python
def getLabels(data):
    immunizationLabels=[]
    informationLabels=[]
    practiceLabels=[]
    for index, row in data.iterrows():
        immunizationStatus="low"
        practiveStatus="bad"
        informationStatus="no"
        informationCount=0
        immunizationCount=0  
        practiceCount=0
        if(row['h0']=='yes'):
            immunizationCount+=1
        if(row['h2']=='yes'):
            immunizationCount+=1
        if(row['h4']=='yes'):
            immunizationCount+=1
        if(row['h6']=='yes'):
            immunizationCount+=1
        if(row['h8']=='yes'):
            immunizationCount+=1
        if(row['h9']=='yes'):
            immunizationCount+=1
        if(row['h9a']=='yes'):
            immunizationCount+=1
        if(row['h51']=='yes'):
            immunizationCount+=1
        if(row['h52']=='yes'):
            immunizationCount+=1
        if(row['h53']=='yes'):
            immunizationCount+=1
        if(row['h54']=='yes'):
            immunizationCount+=1
        if(row['h55']=='yes'):
            immunizationCount+=1
        if(row['h56']=='yes'):
            immunizationCount+=1
        #Immunization Labels
        if(immunizationCount>0):
            immunizationStatus='partial low'
        if(immunizationCount>5):
            immunizationStatus='medium'
        if(immunizationCount>10):
            immunizationStatus='partial high'
        if(immunizationCount==13):
            immunizationStatus='high'
        immunizationLabels.append(immunizationStatus)
        
        #Access to Information Label
        if(row['v157']=='yes'):
            informationCount+=1
        if(row['v158']=='yes'):
            informationCount+=1
        if(row['v159']=='yes'):
            informationCount+=1
        if(row['v169a']=='yes'):
            informationCount+=1
        if(row['v171a']=='yes'):
            informationCount+=1
        #Information Label
        if(informationCount>2):
            informationStatus='yes'
        informationLabels.append(informationStatus)
        #Previous Practices Label
        if(row['m1']=='yes'):
            practiceCount+=1
        if(row['m14']=='yes'):
            practiceCount+=1
        if(row['m70']=='yes'):
            practiceCount+=1
        if(row['h1a']=='Hasdocuments'):
            practiceCount+=1
        #Information Label
        if(practiceCount==4):
            practiveStatus='good'
        if(practiceCount<3):
            practiveStatus='average'
        if(practiceCount==0):
            practiveStatus='bad'
        practiceLabels.append(practiveStatus)
    data['immuneStatus'] = immunizationLabels
    data['accesstoInfo'] = informationLabels
    data['previousPractices'] = practiceLabels
```

```python
getLabels(data)
data.to_csv(r'labeledData.csv')
```
